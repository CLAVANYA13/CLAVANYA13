```python
#Step 1: Initialize Slope and Intercept
#Step 2: Initialize Learning Rate and Iterations
#Step 3: Use for loop till end of iteration value
#Step 4: Calculate Y_pred = m*X+c
#Step 5: Using Partial Differential Calculus, arrive at differential values of slope and intercept
#         PD_Slope = (-2/n) * sum(X * (Y - Y_pred))
#        PD_Intercept = (-2/n) * sum(Y - Y_pred)
#Step 6: Then multiply these differential values with the learning rate and subtract it from previous slope and intercept values
#         in order to get new slope and intercept values
#Step 7 : Now, print the new slope and intercept values
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Laxminarayen/Inceptez-Batch-17/main/Day%205%20Linear%20Regression%20Hands%20on/data.csv")
```

```python
df
```

```python
X = df.iloc[:,0]
Y = df.iloc[:,1]
plt.scatter(X,Y)
```

```python
#Initialize slope(m),intercept(c),learning rate(lr),iterations(iter)

m = 0
c = 0
lr = 0.0001
iter_ = 1000
```

```python
#Lets have a look at the line when slope and intercept is 0
Y_pred1 = m*X+c
plt.scatter(X,Y)
plt.plot([min(X),max(X)],[min(Y_pred1),max(Y_pred1)],color='orange')
```

```python
n = float(len(X))
print ("N values is",n)
for i in range(iter_):
    Y_pred2 = m*X + c
    PD_m = (-2/n) * sum(X * (Y - Y_pred2))
    PD_c = (-2/n) * sum(Y - Y_pred2)
    m = m - (lr * PD_m)
    c = c - (lr * PD_c)
    print("Iteration :",i+1)
    print("Slope :",m)
    print("Intercept :",c)
    
print("Final Slope & Intercept value:",m,c)
```
